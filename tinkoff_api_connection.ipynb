# Получение клиента при помощи токена

```python
from tinkoff.invest import Client

TOKEN = ''

with Client(TOKEN) as client:
    print(client.users.get_accounts())
```

# Проверка получения портфеля клиента

```python
with Client(TOKEN) as client:
    portfolio = client.operations.get_portfolio(account_id='2067069449')
    print(portfolio.positions)
```

# Получение инструментов портфеля

```python
with Client(TOKEN) as client:
    portfolio = client.operations.get_portfolio(account_id='2067069449')
    for pos in portfolio.positions:
        print('Isin: ', pos.figi, ' instrument_type: ', pos.instrument_type, ' quantity: ', pos.quantity.units)
```
