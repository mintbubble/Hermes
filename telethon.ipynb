```python
import os
from datetime import datetime
from pytz import timezone
```

```python
base_directory = r'C:\news'
```

```python
def make_archive(data, path):
    with zipfile.ZipFile(path, mode="w", compression=zipfile.ZIP_DEFLATED, compresslevel=9) as zip_file: 
        dumped_JSON: str = json.dumps(data, ensure_ascii=False, indent=4)
        zip_file.writestr("data.json", data=dumped_JSON.encode("utf-8"))
        zip_file.testzip()
```

```python
from telethon.sync import TelegramClient, events
import json
import zipfile

async with TelegramClient('test1', api_id, api_hash) as client:
    for channel_source in channels_source:
        async for message in client.iter_messages(channel_source):
            date = message.date
            text = message.message
            if date > datetime(2024, 6, 1, tzinfo=timezone('UTC')):
                if text:
                    try:
                        channel_id = message.peer_id.channel_id
                        channel_path = os.path.join(base_directory, str(channel_id))
                        if not os.path.exists(channel_path):
                            os.makedirs(channel_path)
                        message_id = message.id
                        file_path = os.path.join(base_directory, str(channel_id), str(message_id) + '.zip')
                        #if not os.path.isfile(file_path):
                        reactions = []
                        if message.reactions != None:
                            reactions = [{'emoticon': r.reaction.emoticon, 'count': r.count} for r in message.reactions.results]
                        comments = 0
                        if message.replies != None:
                            comments = message.replies.comments
                            
                        edit_date = None
                        if message.edit_date != None:
                            edit_date = message.edit_date.isoformat()
                        data = {
                            'id': message_id, 
                            'text': message.message, 
                            'channel_id': message.peer_id.channel_id, 
                            'date': date.isoformat(), 
                            'edit_date': edit_date,
                            'comments': comments, 
                            'views': message.views, 
                            'forwards': message.forwards, 
                            'reactions': reactions
                           }
                        make_archive(data, file_path)
                        print('added: ', file_path)

                    except Exception as ex:
                        print(message)
                        print(ex)
                        break
            else:
                print('start of year')
                break
```
